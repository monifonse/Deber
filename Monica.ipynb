```python
import pandas as pd
import numpy as np
col1 = np.linspace(0, 4, 201)
df = pd.DataFrame(col1, columns=['t'])
df['s'] = df['t'].apply(np.sin)
print(df)
```

```python
df ["s10"] = df["s"].shift(10)
print(df.head(10))
```

```python
df ["d"] = df ["s"] - df ["s10"]
print (df)
```

```python
def is_positive(x):
  if x> 0:
    return 1
  else:
    return 0

df['p'] = df['d'].apply(lambda x: is_positive(x)) 
df
```

```python
df.to_csv('my_df.csv',sep = ';')
```

```python
df['m'] = df.groupby('p')['d'].transform('mean')
```

```python
# Declaramos la funcion de numero primo 
def primo(numero):
  if numero == 0 or numero == 1 or numero == 4:
    return False
  for x in range (2, int(numero/2)):
    if numero % x == 0:
      return False
  return True
# Utilizamos la sentencia while mientras encuentra los 73 numeros no primos
i = 0
j = 0
while True:
      es_primo = primo(i)
      if es_primo:
          i+=1
      else:
         print("No es primo",i)
         i+= 1
         j+= 1
      if j == 73:
       break
```
